```python
#1. Tampilkan 10 Global_Sales teratas berdasarkan name, genre, platform
game_sales = (
    df.groupby(["Name", "Genre", "Publisher"])["Global_Sales (millions_copy)"]
    .sum()
    .reset_index()
    .sort_values(by="Global_Sales (millions_copy)", ascending=False)
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)

game_sales.head(10)
```

```python
#2. Tampilkan 10 platfrom yang paling banyak digunakan
platform_game_count = (
    df.groupby("Platform")["Name"]
    .nunique()
    .reset_index(name="Jumlah_Video_Game")
    .sort_values(by="Jumlah_Video_Game", ascending=False)
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)

platform_game_count.head(10)
```

```python
#3. Tampilkan berapa banyak video game di setiap genre
genre_game_count = (
    df.groupby("Genre")["Name"]
    .nunique()
    .reset_index(name="Jumlah_Video_Game")
    .sort_values(by="Jumlah_Video_Game", ascending=False)
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)

genre_game_count
```

```python
#4. Tampilkan 10 video game yang terbanyak terjual di wilayah Amerika Utara berdasarkan nama dan genre
top_10_na = (
    df.groupby(["Name", "Genre"])["NA_Sales (millions_copy)"]
    .sum()
    .reset_index(name="Total_NA_Sales")
    .sort_values(by="Total_NA_Sales", ascending=False)
    .head(10)
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)

top_10_na
```

```python
#5. Tampilkan 10 video game yang terbanyak terjual di wilayah Eropa berdasarkan nama dan genre
top_10_eu = (
    df.groupby(["Name", "Genre"])["EU_Sales (millions_copy)"]
    .sum()
    .reset_index(name="Total_EU_Sales")
    .sort_values(by="Total_EU_Sales", ascending=False)
    .head(10)
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)

top_10_eu
```

```python
#6. Tampilkan 10 video game yang terbanyak terjual di wilayah Amerika Utara berdasarkan nama dan genre
top_10_jp = (
    df.groupby(["Name", "Genre"])["JP_Sales (millions_copy)"]
    .sum()
    .reset_index(name="Total_JP_Sales")
    .sort_values(by="Total_JP_Sales", ascending=False)
    .head(10)
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)

top_10_jp
```

```python
#7.Tampilkan 10 teratas dari total video game yang dikeluarkan publisher
publisher_game_count = (
    df.groupby("Publisher")["Name"]
    .nunique()
    .reset_index(name="Total_Video_Game")
    .sort_values(by="Total_Video_Game", ascending=False)
    .head(10)
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)

publisher_game_count
```

```python
#9. Tampilkan Global_Sales setiap tahunnya, urutkan dari yang terbesar
yearly_sales = (
    df.groupby('Year')['Global_Sales (millions_copy)']
    .sum()
    .reset_index()
    .sort_values(by='Global_Sales (millions_copy)',ascending=False)
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)
yearly_sales
```

```python
#11. Tampilkan persentase penjualan tiap wilayah terhadap penjualan global
#Step 1. Menghitung total penjualan untuk setiap wilayah dan global
total_na = df['NA_Sales (millions_copy)'].sum()
total_eu = df['EU_Sales (millions_copy)'].sum()
total_jp = df['JP_Sales (millions_copy)'].sum()
total_other = df['Other_Sales (millions_copy)'].sum()
total_global = df['Global_Sales (millions_copy)'].sum()

#Step 2. Membuat DataFrame baru untuk menampilkan hasil persentase
persentase_wilayah = (
    pd.DataFrame({
        "Wilayah": ["Amerika Utara (NA)", "Eropa (EU)", "Jepang (JP)", "Lainnya (Other)"],
        "Total Penjualan (Millions Copy)": [total_na, total_eu, total_jp, total_other],
        "Penjualan Per Wilayah Terhadap Penjualan Global (%)": [
            (total_na / total_global) * 100,
            (total_eu / total_global) * 100,
            (total_jp / total_global) * 100,
            (total_other / total_global) * 100
        ]
    })
    .sort_values(by="Penjualan Per Wilayah Terhadap Penjualan Global (%)", ascending=False) # Mengurutkan dari persentase terbesar
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)

#Step 3. Menampilkan hasil
persentase_wilayah
```

```python
#12. Tampilkan rentang tahun platform tersebut digunakan
rentang_tahun_platform = (
    df.groupby("Platform")["Year"]
    .agg(Dari_Tahun="min", Hingga_Tahun="max")
    .reset_index()
    .sort_values(by="Dari_Tahun") #Diurutkan dari konsol paling jadul
    .reset_index(drop=True) #Merapikan nomor indeks tabel
)

rentang_tahun_platform
```

```python
#13. Kategorikan platform berdasarkan perkembangan teknologi
def kategori_platform(platform):

    retro = ['NES', 'SNES', 'GB', 'GBA']
    era_3d = ['PS', 'N64', 'DC', 'SAT']
    modern = ['PS2', 'Xbox', 'GC']
    hd = ['PS3', 'X360', 'Wii']
    handheld = ['DS', 'PSP', '3DS']

    if platform in retro:
        return 'Retro Era'

    elif platform in era_3d:
        return '3D Era'

    elif platform in modern:
        return 'Modern Console Era'

    elif platform in hd:
        return 'HD Gaming Era'

    elif platform in handheld:
        return 'Handheld Era'

    else:
        return 'Other'

df['Tech_Category'] = df['Platform'].apply(kategori_platform)
df['Tech_Category'].value_counts()
```

```python
#14. Tampilkan genre yang populer berdasarkan jumlah total penjualan global di setiap tahunnya
genre_populer_per_tahun = (
    df.groupby(["Year", "Genre"])["Global_Sales (millions_copy)"]
    .sum()
    .reset_index(name="Total_Global_Sales")
    #Step 1: Mengurutkan berdasarkan tahun (naik) dan penjualan (turun)
    .sort_values(by=["Year", "Total_Global_Sales"], ascending=[True, False])
    #Step 2: Mengambil baris pertama (penjualan tertinggi) untuk setiap tahun
    .groupby("Year")
    .head(1)
    .reset_index(drop=True)
)

genre_populer_per_tahun
```

```python
#15. Tampilkan top genre setiap publisher berdasarkan total penjualan global
top_genre_publisher_global = (
    df.groupby(["Publisher", "Genre"])["Global_Sales (millions_copy)"]
    .sum()
    .reset_index(name="Total_Global_Sales")
    #Step 1: Cari genre tertinggi di dalam masing-masing publisher
    .sort_values(by=["Publisher", "Total_Global_Sales"], ascending=[True, False])
    .groupby("Publisher").head(1)
    #Step 2: Urutkan list publisher tersebut berdasarkan penjualan terbanyak
    .sort_values(by="Total_Global_Sales", ascending=False)
    .reset_index(drop=True)
)

top_genre_publisher_global
```
