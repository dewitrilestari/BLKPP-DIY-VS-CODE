```python
#Mengubah header kolom sales ke satuan juta copy
data.rename(columns={'NA_Sales': 'NA_Sales (millions_copy)'}, inplace=True)
data.rename(columns={'EU_Sales': 'EU_Sales (millions_copy)'}, inplace=True)
data.rename(columns={'JP_Sales': 'JP_Sales (millions_copy)'}, inplace=True)
data.rename(columns={'Other_Sales': 'Other_Sales (millions_copy)'}, inplace=True)
data.rename(columns={'Global_Sales': 'Global_Sales (millions_copy)'}, inplace=True)
```

```python
# 5. video game nama, genre yang banyak terjual di wilayah eropa
data[["Name", "Genre", "EU_Sales (millions_copy)"]].sort_values(by="EU_Sales (millions_copy)", ascending=False)
```

```python
# 6. video game nama, genre yang banyak terjual di wilayah jepang
data[["Name", "Genre", "JP_Sales (millions_copy)"]].sort_values(by="JP_Sales (millions_copy)", ascending=False)
```

```python
# Melihat persentase tiap negara terhadap penjualan global
total_sales_NA = data['NA_Sales (millions_copy)'].sum()
total_sales_EU = data['EU_Sales (millions_copy)'].sum()
total_sales_JP = data['JP_Sales (millions_copy)'].sum()
total_sales_Other = data['Other_Sales (millions_copy)'].sum()
total_sales_Global = data['Global_Sales (millions_copy)'].sum()
```

```python
# Percentase penjualan di amerika
percent_sales_NA = (total_sales_NA / total_sales_Global) * 100
print(f"NA Sales: {percent_sales_NA:.2f}%")
```

```python
# Percent penjualan di eropa
percent_sales_EU = (total_sales_EU / total_sales_Global) * 100
print(f"EU Sales: {percent_sales_EU:.2f}%")
```

```python
# percent sales di Jepang
JP = (total_sales_JP / total_sales_Global) * 100
print(f"JP Sales: {JP:.2f}%")
```

```python
# Percent other sales
other_sales = (total_sales_Other / total_sales_Global) * 100
print(f"other Sales: {other_sales:.2f}%")
```

```python
# (Tambahan) Groupby tahun, sum penjualan global
yearly_sales = data.groupby('Year')['Global_Sales (millions_copy)'].sum().reset_index().sort_values(by='Global_Sales (millions_copy)',ascending=False)

yearly_sales
```

```python
#7.Berapa total Video Game yang sudah dikeluarkan publisher
data.groupby(["Publisher"])["Name"].count().sort_values(ascending=False).head(10)
```

```python
#8.Jumlah penjualan video game di Amerika Utara, Eropa, dan Jepang
data["Penjualan di Amerika Utara, Eropa, dan Jepang (millions_copy)"] = data["Global_Sales (millions_copy)"] - data["Other_Sales (millions_copy)"]

data[["Name", "Penjualan di Amerika Utara, Eropa, dan Jepang (millions_copy)"]].head(10)
```
